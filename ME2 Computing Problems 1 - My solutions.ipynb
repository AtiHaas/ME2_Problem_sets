```python
#TASK A 

import numpy as np
import math as mt
import matplotlib.pyplot as pl

#Defining the domains and discretiseing them:
x1 = np.arange(-5,-1.5,0.5)
x2 = np.arange(-2.05,3, 0.05)
x3 = np.arange(3.5,5,0.5)

#Creating the arrays for the target values:
y1 = []
y2 = []
y3 = []


#Calculating the values for y:
for x in x1:
    y1.append(np.sin(x))
for x in x2:
    y2.append(np.sin(x))
for x in x3:
    y3.append(np.sin(x))

#Plotting the results:
pl.scatter(x1,y1)
pl.scatter(x2,y2)
pl.scatter(x3,y3)
```

```python
#TASK B
from mpl_toolkits import mplot3d


#Defining the domains and discretiseing them:
x = np.arange(-2*np.pi, 2*np.pi+0.1, 0.1)
y = np.arange(-1*np.pi, 2*np.pi+0.1, 0.1)

#Creating the grids:
(Xg, Yg) = np.meshgrid(x,y) #This function creates the 2d grids as needed
#F = np.ndarray((Ny,Nx))     #This function creates the targer array

#Calculating the values in each point of the grid:
F = np.sin(Xg) * np.cos(Yg) #Here I compute the values for the target array, IMPORTANT: Use numpy library for the math functions (Only this will do it mesh wise)
G = np.cos(Xg) * np.sin(Yg)
S = F + G
P = F * G

# print sizes to examine
print(len(x))
print(len(y))
print(np.shape(Xg))
print(np.shape(Yg))
```

```python
#TASK C PART 1

#Displaying the results:

#Plotting S
print('S(x,y) is the following in 3D:')
ax = pl.axes(projection='3d')
ax.plot_surface(Xg,Yg,S)


#Plotting the contour of S
pl.show()           # make a new window plot
print('S(x,y) has the following contour plot:')
pl.contour(Xg,Yg,S) # plot contours

#Plotting P 
pl.show()           # make a new window plot
print('P(x,y) is the following in 3D:')
ax = pl.axes(projection='3d')
ax.plot_surface(Xg,Yg,P)

#Plotting the contour of P
pl.show()           # make a new window plot
print('P(x,y) has the following contour plot:')
pl.contour(Xg,Yg,P) # plot contours



```

```python
#TASK C PART 2 AND 3

#Creating and discretiseing the time domain:
dt = 0.05
t = np.arange(0, 10+dt, dt)


#Creating the grids
(Xg, Yg, Tg) = np.meshgrid(x, y, t)


#Calculating the values for each mesh 
R = np.sin(Xg) * np.cos(Yg) * np.exp(-0.5*Tg)

#Plotting R at t = 0
print('The 3D plot of R at t=0s will be:')
ax = pl.axes(projection='3d')
ax.plot_surface(Xg[:,:,0], Yg[:,:,0], R[:,:,0])
    
    
#Plotting R at t = 5
pl.show() 
print('The 3D plot of R at t=5s will be:')
ax = pl.axes(projection='3d')
ax.plot_surface(Xg[:,:,5], Yg[:,:,5], R[:,:,5])
```

```python
#TASK D 

#Creating and discretiseing the domains:
dx = 0.5
dy = 0.5
x = np.arange(-5, 5+dx, dx)
y = np.arange(-5, 5+dy, dy)

LenX = len(x)
LenY = len(y)

Xg, Yg = np.meshgrid(x,y)

# allocate an array for F: size is Nx by Ny by 2 (2 is because the vector has two components, i and j)
F_a = np.ndarray((LenX,LenY, 2))
F_b = np.ndarray((LenX,LenY, 2))
F_c = np.ndarray((LenX,LenY, 2))

#Calculating the values:
F_a[:,:,0] = Xg #WATCH OUT THE INDEXING IS FROM 0
F_a[:,:,1] = Yg

F_b[:,:,0] = Yg
F_b[:,:,1] = -Xg

F_c[:,:,0] = Yg / (Xg ** 2 + Yg ** 2)
F_c[:,:,1] = -Xg / (Xg ** 2 + Yg ** 2)

#Plotting the values:
#Problem a)
print('The vector field in Problem a), plotted, looks like the following:')
pl.quiver(Xg, Yg, F_a[:,:,0], F_a[:,:,1])

pl.show()
print('The vector field in Problem a), plotted as streamlines, looks like the following:')
pl.streamplot(Xg, Yg, F_a[:,:,0], F_a[:,:,1])
pl.show()

#Problem b)
print('The vector field in Problem a), plotted, looks like the following:')
pl.quiver(Xg, Yg, F_b[:,:,0], F_b[:,:,1])

pl.show()
print('The vector field in Problem a), plotted as streamlines, looks like the following:')
pl.streamplot(Xg, Yg, F_b[:,:,0], F_b[:,:,1])
pl.show()

#Problem c)
print('The vector field in Poroblem a), plotted, looks like the following:')
pl.quiver(Xg, Yg, F_c[:,:,0], F_c[:,:,1])

pl.show()
print('The vector field in Poroblem a), plotted as streamlines, looks like the following:')
pl.streamplot(Xg, Yg, F_c[:,:,0], F_c[:,:,1])
```
